```python
import pandas
import numpy
df = pandas.read_csv('In_march_to_oct2562.csv', low_memory=False)
df
```

```python
typeName = df['typ_name'].unique()
month = df['month'].unique()

result = dict()

for t in typeName:
    r_m = []
    for m in month:
        sumCost = sum(df[(df['typ_name'] == t) & (df['month'] == m)]['cost_build'])
        
        r_m.append(sumCost)
    result[t] = r_m
pandas.DataFrame(result)
```
